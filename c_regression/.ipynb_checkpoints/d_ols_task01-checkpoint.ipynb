```python
pre_c_df['target'] = pre_c_df['Spending Score (1-100)']
pre_c_df = pre_c_df.drop(labels=['Spending Score (1-100)'], axis=1).reset_index(drop=True)
pre_c_df
```

```python
from statsmodels.api import OLS

features, targets = pre_c_df.iloc[:, 1:], pre_c_df.iloc[:, 0]

model = OLS(targets, features)
print(model.fit().summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def feature_engineering_VIF(features):
    vif = pd.DataFrame()
    vif['score'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
    vif['feature'] = features.columns
    return vif
```

```python
display(feature_engineering_VIF(features))
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

c_df = c_df.copy()
columns = ['Profession']

for column in columns:
    one_hot_encoder = OneHotEncoder(sparse_output=False)
    result = one_hot_encoder.fit_transform(c_df[[column]])

    c_df = pd.concat([c_df, 
                          pd.DataFrame(result, columns=one_hot_encoder.categories_).astype(np.int8)], 
                         axis=1)
```

```python
c_df = c_df[~ c_df['Profession'].isna()]
c_df.isna().sum()
```
